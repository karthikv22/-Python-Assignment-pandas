```python
#  59.Convert the OrderDate to datetime and extract new columns: Year, Month, and Weekday.
import pandas as pd

sales['OrderDate'] = pd.to_datetime(sales['OrderDate'], errors='coerce')

sales['Year'] = sales['OrderDate'].dt.year
sales['Month'] = sales['OrderDate'].dt.month
sales['Weekday'] = sales['OrderDate'].dt.day_name()
```

```python
# 60.Create a monthly revenue trend using resample() on OrderDate.
sales = sales.sort_values('OrderDate')
sales = sales.set_index('OrderDate')

monthly_revenue = sales['TotalAmount'].resample('M').sum().reset_index()

print(monthly_revenue.head())
```

```python
# 61. Find the day of the week with the highest sales volume.
sales['Weekday'] = sales.index.day_name()

weekday_revenue = sales.groupby('Weekday')['TotalAmount'].sum().sort_values(ascending=False)
print(weekday_revenue)

top_weekday_revenue = weekday_revenue.idxmax()
print("Day with highest revenue:", top_weekday_revenue)
```

```python
# 63. Find peak sales month and peak returned month for each Region.
sales = sales.reset_index()
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'], errors='coerce')
sales['YearMonth'] = sales['OrderDate'].dt.to_period('M')  # e.g. 2024-01

monthly_region = (sales
    .groupby(['Region', 'YearMonth'])
    .agg(
        total_sales=('TotalAmount', 'sum'),
        returned_orders=('IsReturned', 'sum')
    )
    .reset_index()
)
```
